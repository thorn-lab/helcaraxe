```python
import pandas as pd
```

```python
pkl_path = "/Users/kristophernolte/Documents/GitHub/helcaraxe/pdb_analysis/PDB_REPORT_pkl/train_set_meta"
main_df = pd.read_pickle(pkl_path+".pkl")
main_df.head()
```

```python
main_df = main_df.drop(main_df.index[0])
```

```python
main_df = main_df.shift(periods=1, axis="columns")
main_df['Entry ID'] = main_df.index
```

```python
main_df["Collection Temperature"] = main_df["Collection Temperature"].astype("float64")
main_df["Matthews Coefficient"] = main_df["Matthews Coefficient"].astype("float64")
main_df["Percent Solvent Content"] = main_df["Percent Solvent Content"].astype("float64")
main_df["pH"] = main_df["pH"].astype("float64")
main_df["Temp (K)"] = main_df["Temp (K)"].astype("float64")
main_df["Average B Factor"] = main_df["Average B Factor"].astype("float64")
main_df["R Free"] = main_df["R Free"].astype("float64")
main_df["R Work"] = main_df["R Work"].astype("float64")
main_df["R All"] = main_df["R All"].astype("float64")
main_df["R Observed"] = main_df["R Observed"].astype("float64")
main_df["High Resolution Limit"] = main_df["High Resolution Limit"].astype("float64")
main_df["Reflections For Refinement"] = main_df["Reflections For Refinement"].astype("float64")
main_df.shape
main_df.to_pickle(pkl_path+"_clean.pkl")
```
